```python
import json
import pandas as pd

# Define the file path
file_path = 't20_json_files/t20_json_files/t20_wc_batting_summary.json'

# Ensure the file exists before attempting to open it
try:
    with open(file_path) as f:
        data = json.load(f)
        
        all_records = []
        for rec in data:
            if 'battingSummary' in rec:  # Check if 'battingSummary' key exists
                all_records.extend(rec['battingSummary'])
        
        df_batting = pd.DataFrame(all_records)
        
        # Print the first few rows of the DataFrame
        print(df_batting.head())

except FileNotFoundError:
    print(f"The file at {file_path} was not found. Please check the path and try again.")
except json.JSONDecodeError:
    print(f"Failed to decode JSON from the file at {file_path}. Ensure the file contains valid JSON.")
except KeyError as e:
    print(f"The expected key {e} was not found in the JSON data. Please check the structure of your JSON file.")
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) >0 else"not out")
df_batting.head(11)
```

```python
df_batting.drop(columns = ["dismissal"],inplace = True)
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

# Print the match_ids_dict to verify
print(match_ids_dict)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)

df_batting.head()
```

```python
df_batting.to_csv('t20_json_files/t20_json_files/batting.csv', index=False)
```
